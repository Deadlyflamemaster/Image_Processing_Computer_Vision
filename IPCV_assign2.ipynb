```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
inp = int(input('Which image to display?'))
```

```python
if(inp==1): path = '/content/gutters1.JPG'
if(inp==2): path = '/content/gutters2.JPG'
if(inp==3): path = '/content/gutters3.JPG'
```

```python
image = cv2.imread(path)

arr_planes = cv2.split(image)
arr_processed_planes = []

for plane in arr_planes:
    dilated = cv2.dilate(plane, np.ones((11,11), np.uint8))
    blurred = cv2.medianBlur(dilated, 19)
    diff = 255 - cv2.absdiff(plane, blurred)
    # norm = cv2.normalize(diff,None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)
    arr_processed_planes.append(diff)

processed_img = cv2.merge(arr_processed_planes)

# Display the processed image
cv2_imshow(processed_img)
```
