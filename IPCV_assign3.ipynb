```python
#Imports
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Initialize blank display
img = np.zeros((300, 500), dtype=np.uint8)
# The center coordinates of all 15 circles for a digit are stored.
circles = []
x=25
y=50
for j in range(3):
  y=50;
  for i in range(5):
    circles +=  [[x,y]]
    y+=50
  x+=50
shift = 290
```

```python
#Input
N = int(input("Enter your Roll Number: "))%100
```

```python
# For each digit, generate their image.
while(N):
  digit = N%10
  N = (int)(N/10)

  if(digit==0):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [6,7,8]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==1):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [0,1,2,3,4,10,11,12,13,14]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==2):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [1,6,13,8]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==3):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [1,6,3,8]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==4):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [5,6,3,4,8,9]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==5):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [6,11,3,8]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==6):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [6,11,8]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==7):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [1,2,3,4,6,7,8,9]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift  
  if(digit==8):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [6,8]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  if(digit==9):
    for i in range(len(circles)):
      circles[i][0] = circles[i][0] + shift
      if i not in [6,3,8]:
        cv2.circle(img, tuple(circles[i]), 25, 255, -1);
      circles[i][0] = circles[i][0] - shift
  shift = 60;
```

```python
# Show the image as a grayscale image.
plt.imshow(img, 'gray')
```
